```python
  # Library for countries


import pycountry

# Gets list of countries
listOfCountries = []
for x in pycountry.countries:
      listOfCountries.append(x.official_name if hasattr(x, "official_name") else x.name)
```

```python
# Gets frequency of countries in df_projects_healthcare, filters by countries (single country, name match to a country in listOfCountries)
tempCountriesFreq = df_projects_healthcare['Country'].value_counts().to_dict()
countriesFreq = {k: tempCountriesFreq[k] for k in listOfCountries if k in tempCountriesFreq}   

countriesFreqAsDF = pd.DataFrame(countriesFreq.items(), columns=["country", "count"])

countriesFreqAsDF.sort_values(by="count", ascending=False).head(10)
```

```python
# FUNCTION TO GET THE ALPHA-3-CODE OF EACH COUNTRY -- https://melaniesoek0120.medium.com/data-visualization-how-to-plot-a-map-with-geopandas-in-python-73b10dcd4b4b
def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(official_name=country).alpha_3
            
           # .alpha_3 means 3-letter country code 
           # .alpha_2 means 2-letter country code
            CODE.append(code)
        except:
            code=pycountry.countries.get(name=country).alpha_3
            CODE.append(code)
    return CODE
```

```python
# Creates col "CODE" with alpha3code
countriesFreqAsDF["CODE"] = alpha3code(countriesFreqAsDF.country)
```

```python
# 'naturalearth_lowres' is geopandas datasets so we can use it directly
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))

# rename the columns so that we can merge with our data
world.columns=['pop_est', 'continent', 'name', 'CODE', 'gdp_md_est', 'geometry']

# then merge with our data 
merge=pd.merge(world,countriesFreqAsDF,on='CODE', how="outer")

# last thing we need to do is - merge again with our location data which contains each country’s latitude and longitude
location=pd.read_csv('https://raw.githubusercontent.com/melanieshi0120/COVID-19_global_time_series_panel_data/master/data/countries_latitude_longitude.csv')
merge=merge.merge(location,on='name').sort_values(by='count',ascending=False).reset_index()
```

```python
# Sets any Nan count values to 0, and plots the graph
merge["count"] = merge["count"].fillna(0)
merge.plot(column='count', legend=True,cmap='Blues', legend_kwds={'label': "Distribution of Projects", 'orientation': "horizontal"})
```

```python
#Hypothesis Test:  Two-Sample Z-test for difference in average project cost of healthcare projects based on project status
from scipy import stats

costGivenActive = df_projects_healthcare[df_projects_healthcare["Project Status"]=="Active"]["Current Project Cost"].dropna()
#print(costGivenActive)
costGivenNotActive = df_projects_healthcare[df_projects_healthcare["Project Status"]!="Active"]["Current Project Cost"].dropna()
#print(costGivenNotActive)

twoSidedtTest = stats.ttest_ind(costGivenActive, costGivenNotActive, axis=0)
qHypothesis = twoSidedtTest.pvalue # p-value
print(qHypothesis)
print('''Two-Sample Z-test
H0: 𝜇1−𝜇2=0. (No difference in average Project Cost of healthcare projects based on Active or not active Project Status)

HA: 𝜇1−𝜇2 ≠ 0 (Some difference in average Project Cost of healthcare projects based on Active or not active Project Status)

Because the p-value of our test (0.013391877332543005) is way less than alpha = 0.05, we reject the null hypothesis of the test in favor of the alternate hypothesis. 
We have sufficient evidence to say that there is a significant difference in project cost of healthcare projects based on project status (whether the project is active or not active).
''')
```

```python
data = df_projects_healthcare_ml['Project Development Objective ']
target = df_projects_healthcare_ml['Environmental and Social Risk']
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_target, test_target = train_test_split(data, target, 
                                                                    test_size=0.3, 
                                                                    random_state=216)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.metrics import plot_confusion_matrix

nb_model= make_pipeline(CountVectorizer(binary=True),MultinomialNB())
nb_model.fit(train_data,train_target)
prediction= nb_model.predict(test_data)
acc = accuracy_score(test_target, prediction)
print(acc)

plot_confusion_matrix(nb_model,data,target)
plt.grid(False)
```
